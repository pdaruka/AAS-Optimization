```python
query = """
SELECT buy.Origin,
buy.Destination,
buy.Container_Size,
originport.Port_Code AS Origin_Port,
destinationport.Port_Code AS Destination_Port,
CASE 
WHEN buy.Carrier='Hapag-Lloyd' THEN 'HLCU'
WHEN buy.Carrier='CMA' THEN 'CMDU'
WHEN buy.Carrier='Maersk' THEN 'MAEU'
ELSE buy.Carrier
END AS Carrier_Name,
CASE 
WHEN Notes LIKE 'Bunker Included%' THEN Freight_Rate
ELSE Freight_Rate+Bunker_Rate 
END AS Rate
FROM `wf-gcp-us-ae-ops-prod.csn_reporting_isc.tbl_isc_pricing_ocean_buy_rates` buy
LEFT JOIN `wf-gcp-us-ae-ops-prod.csn_reporting_isc.tbl_aas_portmapping` originport ON buy.Origin=originport.Port
LEFT JOIN `wf-gcp-us-ae-ops-prod.csn_reporting_isc.tbl_aas_portmapping` destinationport ON buy.Destination=destinationport.Port
WHERE Container_Size NOT IN ('20DRY','45HDRY','20HDRY')
AND (Oc_Effective_Date<= CURRENT_DATE() OR Oc_Effective_Date IS NULL)
AND (Oc_Expiry_Date >=CURRENT_DATE() OR Oc_Expiry_Date IS NULL)
GROUP BY 1,2,3,4,5,6,7
"""
```

```python
ocean_rate = EasyQuery(cmd=query, target='eden-gbq-ops-prod').get_dataframe()
```

```python
from datetime import timedelta
```

```python
#specifying priority of carrier
from pandas.api.types import CategoricalDtype
data['carrier'] = data['carrier'].astype(pd.CategoricalDtype(
                  ['HLCU','CMDU','MAEU'], ordered=True))
```

```python
#removing duplicates
data.drop_duplicates(inplace=True)
```

```python
#replacing all occurences of VNVUT with VNSGN as they are interchangeable
data['origin'] = data['origin'].str.replace('VNVUT', 'VNSGN')
data['destination'] = data['destination'].str.replace('VNVUT', 'VNSGN')
df_pendingreport['Origin'] = df_pendingreport['Origin'].str.replace('VNVUT', 'VNSGN')
df_pendingreport['Destination'] = df_pendingreport['Destination'].str.replace('VNVUT', 'VNSGN')
```

```python
df_pendingreport['ECRD']=pd.to_datetime(df_pendingreport['ECRD'])
data['etd']=pd.to_datetime(data['etd'])
data['eta']=pd.to_datetime(data['eta'])
```

```python
#converting dates to correct format
date1 = pd.to_datetime(data['eta'], errors='coerce', format='%Y-%d-%m %H:%M:%S').astype('datetime64[D]')
date2 = pd.to_datetime(data['eta'], errors='coerce', format='%m/%d/%Y')
data['eta'] = date1.fillna(date2)
data['eta']=pd.to_datetime(data['eta'])
```

```python
date1 = pd.to_datetime(data['etd'], errors='coerce', format='%Y-%d-%m %H:%M:%S').astype('datetime64[D]')
date2 = pd.to_datetime(data['etd'], errors='coerce', format='%m/%d/%Y')
data['etd'] = date1.fillna(date2)
data['etd']=pd.to_datetime(data['etd'])
```

```python
data['available_capacity']=pd.to_numeric(data['available_capacity'])
data['allocated_capacity']=pd.to_numeric(data['allocated_capacity'])
```

```python
cols=df_pendingreport.columns
cols
```

```python
# initializing dataframe which holds final result
fin=pd.DataFrame()
```

```python
data.dtypes
```

```python
df_pendingreport.dtypes
```

```python
for ind in range(len(df_pendingreport)):
  orig=df_pendingreport.loc[ind,'Origin'] #origin in pending report
  dest=df_pendingreport.loc[ind,'Destination'] #destination in pending report
  ecrd=df_pendingreport.loc[ind,'ECRD'] #ECRD in pending report
  min_etd=ecrd+timedelta(days=5)#min time buffer
  max_etd=ecrd+timedelta(days=20)#max time buffer
  df1=data[(data.origin==orig) & (data.destination==dest) & (data.etd>=min_etd) & (data.etd<=max_etd) & (data.available_capacity>0)]# filtering rows which satisfy conditions
  df1.reset_index(inplace=True,drop=True)
  df1.sort_values(by=['carrier','available_capacity'],ascending=[True,False],inplace=True)#sorting first by type of carrier then by decreasing capacity
  df1.reset_index(inplace=True,drop=True)
  df1['rank_cap'] = df1.index + 1 #rank acc to capacity
  df1.sort_values(['rate','carrier'],inplace=True)#sorting first by cost then by type of carrier
  df1.reset_index(inplace=True,drop=True)
  df1['rank_cost'] = df1.index + 1 #rank acc to cost
  df1.sort_values(['eta','carrier'],ascending=[True,True],inplace=True)#sorting first by eta then by type of carrier
  df1.reset_index(inplace=True,drop=True)
  df1['rank_time'] = df1.index + 1 #rank acc to time of arrival
  s=df_pendingreport.loc[[ind],cols]#the current row of pending report as a dataframe
  df2=s.merge(df1, how='outer',left_on='Origin',right_on='origin')#joining current row with all service string allocations with the required ranks
  fin=pd.concat([fin,df2],ignore_index=True)#adding them to final dataset
```
